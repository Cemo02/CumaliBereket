```python
import pandas as pd
import numpy as np
import random
from datetime import datetime
```

```python
# Read the Excel file into a DataFrame
receipts = pd.read_excel('receipts.xlsx')
```

```python

# Define a list of new columns
new_columns = ['Opinion on Business Customer', 'Rate business customer (from 1 to 5)',
               'Selected product from record', 'Opinion on one selected product',
               'Rate on one selected product (from 1 to 5)']

# Generate new client data for each unique "Rachunek" value
new_data = pd.DataFrame(columns=['ID of Business Customer', 'Business customer name'] + new_columns)

# Dictionary to store generated IDs and names for each 'Rachunek'
generated_data_dict = {}

for index, row in receipts.iterrows():
    rachunek = row['Rachunek']

    # Check if the current 'Rachunek' already exists in the new_data DataFrame
    if rachunek not in generated_data_dict:
        # Generate unique ID and name for the 'Rachunek'
        generated_id = np.random.randint(1, 100000)
        generated_name = f'La Cedre {np.random.randint(1, 100)}'

        # Add the generated ID and name to the dictionary
        generated_data_dict[rachunek] = {'ID of Business Customer': generated_id, 'Business customer name': generated_name}

        # Append the new client data to the new_data DataFrame
        new_data = new_data.append(generated_data_dict[rachunek], ignore_index=True)

    # Update the corresponding row in the receipts DataFrame with the generated client data
    receipts.at[index, 'ID of Business Customer'] = generated_data_dict[rachunek]['ID of Business Customer']
    receipts.at[index, 'Business customer name'] = generated_data_dict[rachunek]['Business customer name']

# Initialize the new columns with empty values
for col in new_columns:
    receipts[col] = ''

# Reorder the columns in the receipts DataFrame
receipts = receipts[['ID of Business Customer', 'Business customer name'] + receipts.columns[:-2].tolist() + new_columns]

# Drop the 'Klient' column from receipts DataFrame
receipts.drop(columns='Klient', inplace=True)

# Print the updated DataFrame
print(receipts)
```
