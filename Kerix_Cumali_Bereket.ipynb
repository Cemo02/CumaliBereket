```python
import pandas as pd
import numpy as np
import random

# Read the Excel file into a DataFrame
receipts = pd.read_excel('receipts.xlsx')

# Read the name list DataFrame from 'name_list.xlsx' (assuming it has a 'name' column)
name_list = pd.read_excel('clients_data.xlsx', sheet_name='first_name')

```

```python
name_list
```

```python
# Generate new client data for each unique "Rachunek" value
new_data = pd.DataFrame(columns=['Rachunek', 'Klient', 'age', 'sex', 'email', 'status'])

for index, row in receipts.iterrows():
    rachunek = row['Rachunek']

    # Check if the current 'Rachunek' already exists in the new_data DataFrame
    if rachunek not in new_data['Rachunek'].values:
        klient = np.random.choice(name_list['first_name'])
        age = np.random.randint(16, 81)
        imei = ''.join(np.random.choice(np.arange(10), size=16).astype(str))
        email = str(klient) + imei + '@cubecloud.com'

        # Assign the status based on the age
        if age <= 24:
            status = '18-24 lat'
        elif age <= 39:
            status = '25-39 lat'
        elif age <= 55:
            status = '40-55 lat'
        else:
            status = '>55 lat'

        #Sex
        name = klient
        lastchar = name[-1]
        if lastchar == 'a':
          sex = random.choices(population = ('female', 'other'), weights = [0.8, 0.2])
        else:
          sex = random.choices(population = ('male', 'other'), weights = [0.8, 0.2])


        # Append the new client data to the new_data DataFrame
        new_data = new_data.append({'Rachunek': rachunek, 'Klient': klient, 'age': age,
                                    'sex': sex, 'email': email, 'status': status}, ignore_index=True)

    # Find the index of the current 'Rachunek' in the new_data DataFrame
    index1 = new_data[new_data['Rachunek'] == rachunek].index[0]

    # Update the corresponding row in the receipts DataFrame with the generated client data
    receipts.at[index, 'Klient'] = new_data.at[index1, 'Klient']
    receipts.at[index, 'age'] = new_data.at[index1, 'age']
    receipts.at[index, 'sex'] = new_data.at[index1, 'sex']
    receipts.at[index, 'email'] = new_data.at[index1, 'email']
    receipts.at[index, 'status'] = new_data.at[index1, 'status']

# Print the updated DataFrame
#print(receipts)
```
